# An√°lise de Produtividade na Constru√ß√£o Civil

A constru√ß√£o civil tem se mostrado uma das √°reas mais rent√°veis e din√¢micas da economia moderna, impulsionada pelo crescimento acelerado da popula√ß√£o em regi√µes urbanas e pela crescente demanda por infraestrutura, habita√ß√£o e servi√ßos. # 3. üìä Mostrando a obra com maior produtividade a partir de gr√°ficos

## Gr√°fico de Densidade

- Densidade por Produtividade, quanto mais alto a Densidade maior a frequencia de dados

- Quanto maior o √≠ndice de produtividade (eixo X), melhor a produtividade

## Gr√°fico de BoxPlot

- Resume a distribui√ß√£o dos dados em termos de quartis (Q1, mediana, Q3) e outliers.

- A mediana (linha dentro da caixa) √© o valor central, mas n√£o reflete necessariamente a concentra√ß√£o de dados.

- No boxplot, a Obra A pode ter uma mediana mais alta do que a Obra B, mesmo que tenha muitos valores de alta produtividade (baixos H/M¬≤).

## !!!!!Pontos importantes
- O gr√°fico de densidade mostra a concentra√ß√£o dos dados, destacando onde h√° mais valores de alta produtividade.

- O boxplot foca na mediana e na dispers√£o dos dados, que pode ser afetada por valores extremos (outliers) ou uma distribui√ß√£o assim√©trica.

```python
# Criando imports para a cria√ß√£o de gr√°ficos de densidade
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")

# Calculando o √≠ndice de produtividade
df['Produtividade'] = df['COEF_SIURB'] / df['IP_D']

plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Produtividade', hue='OBRA', fill=False, common_norm=False, palette='viridis')
plt.title('Densidade de Produtividade por Obra')
plt.xlabel('√çndice de Produtividade (H/M2)')
plt.ylabel('Densidade')
plt.show()
```

```python
produtividade_media = df.groupby('OBRA')['Produtividade'].mean().sort_values(ascending=False)

print("√çndice de Produtividade por Obra:")
display(produtividade_media)

obra_menos_produtiva = produtividade_media.idxmax()
obra_mais_produtiva = produtividade_media.idxmin()

print(f"\nA obra mais produtiva √©: {obra_mais_produtiva} com uma produtividade m√©dia de {produtividade_media.min():.4f} h/m¬≤")
print(f"A obra menos produtiva √©: {obra_menos_produtiva} com uma produtividade m√©dia de {produtividade_media.max():.4f} h/m¬≤")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='OBRA', y='IP_D', data=df, hue='OBRA', palette='viridis', legend=False)
plt.title('Boxplot de √çndice de Produtividade por Obra')
plt.xlabel('Obra')
plt.ylabel('√çndice de Produtividade (H/M¬≤)')
plt.show()
```
