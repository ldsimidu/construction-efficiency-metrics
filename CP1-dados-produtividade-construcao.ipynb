# Análise de Produtividade na Construção Civil

A construção civil tem se mostrado uma das áreas mais rentáveis e dinâmicas da economia moderna, impulsionada pelo crescimento acelerado da população em regiões urbanas e pela crescente demanda por infraestrutura, habitação e serviços. Segundo o relatório **World Urbanization Prospects** da ONU, estima-se que até 2050 cerca de **70% da população global** viverá em cidades, o que coloca o setor em um papel central no desenvolvimento econômico e social. No entanto, esse crescimento também traz desafios significativos, como a necessidade de entregar projetos com maior eficiência, qualidade e dentro dos orçamentos cada vez mais apertados. Nesse cenário, a **produtividade** torna-se um fator crítico para garantir a viabilidade financeira das construções, permitindo uma melhor gestão de recursos, redução de custos e, consequentemente, uma margem de lucro mais atraente.

Para mensurar a eficiência no setor, dois indicadores são especialmente relevantes:

1. **Índice de Produtividade**: Reflete a relação entre os recursos utilizados (mão de obra, materiais e tempo) e os resultados obtidos, identificando oportunidades de melhoria.
2. **Coeficiente do Sistema de Indicadores da Indústria da Construção (SIURB)**: Desenvolvido para avaliar a eficiência na gestão de obras públicas e privadas, oferece uma métrica padronizada para comparar desempenhos e estabelecer padrões de desempenho no setor.

Ambos os indicadores são ferramentas valiosas para entender como a produtividade pode ser otimizada, impactando diretamente o orçamento e o lucro das construções.

## Coletando Dados

Pensando nisso, realizaremos um estudo utilizando **Python** para análise de dados, com o objetivo de entender como o nível de produtividade afeta a escolha dos materiais, local e equipe para uma construção civil. Para isso, um serviço de alvenaria foi acompanhado em **3 obras**:

- **Obra A**
- **Obra B**
- **Obra C**

Todas localizadas no município de **São Paulo**. Nessas obras, foram utilizados **3 tipos de blocos vazados de concreto** com espessuras de **9 cm**, **14 cm** e **19 cm**, respectivamente. As dimensões de comprimento e largura eram iguais para os 3 tipos: **19 cm x 39 cm**.

O indicador de cada dia está denominado como **"IP Diário"**. O recurso de mão de obra foi medido em horas (h) consumidas para a quantidade de produto realizado e o serviço em metros quadrados levantados (m²).

# **1. 🧩Tipos de variáveis:**

- **Coluna sem nome: quantitativa discreta**
     - *São números inteiros e contáveis, responsáveis por enumerar as linhas do arquivo.*

- **Obra: qualitativas nominais**
     - *Embora consigamos ordenar as obras A, B e C de forma alfabética isso não é uma informação relevante para a análise, pois enxergamos essas letras como nomes próprios das obras, mas poderiam ser trocados para algo como: "Galpão Plaza", "Condominio Lagoas" e "Condominio Pitangueiras".*

- **Created: quantitativa contínua**
     - *Ao inserir esses dados em um gráfico, é possível classificá-los em uma linha do tempo com base nas datas registradas e analisar o andamento diário das obras.*

- **Descrição: qualitativa nominal**
     - *É uma variável que apresenta a descrição dos blocos e espessuras e não possui uma ordem definida, já que não tem uma relação de "maior que" ou "menor que", são apenas diferentes.*

- **Classe_Comp: qualitativa nominal**
     - *É usado para classificar e diferenciar os tipos de categorias que cada bloco estão inseridos, servindo como um dado informacional e não ordinal.*  

- **Insumo_Atual: qualitativa nominal**
     - *São dados que informam o tipo de trabalho realizado na obra em questão.*

- **IP_D: quantitativo contínuo**
     - *Os dados estão em decimais com muitas casas após a vírgula, o que traz uma especificidade maior e raramente terá um dado igual ao outro. Portanto, é possível organizar uma linha do tempo que mostre o nível de produtividade das obras ao longo dos dias de uma maneira contínua, em um gráfico, por exemplo.*

- **Unidade: qualitativa nominal**
     - *É uma coluna que informa a unidade de medida que os cálculos de IP foram feitos, logo não é possível ordenar de alguma forma.*

- **COEF_SIURB: qualitativo contínuo**
     - *Assim com a coluna de IP_D, nesta os dados possuem muitas casas decimais, o que tornam os números específicos e abrem margem para um gráfico contínuo que apresente a produtividade ao longo dos dias.*

# 3. 📊 Mostrando a obra com maior produtividade a partir de gráficos

## Gráfico de Densidade

- Densidade por Produtividade, quanto mais alto a Densidade maior a frequencia de dados

- Quanto maior o índice de produtividade (eixo X), melhor a produtividade

## Gráfico de BoxPlot

- Resume a distribuição dos dados em termos de quartis (Q1, mediana, Q3) e outliers.

- A mediana (linha dentro da caixa) é o valor central, mas não reflete necessariamente a concentração de dados.

- No boxplot, a Obra A pode ter uma mediana mais alta do que a Obra B, mesmo que tenha muitos valores de alta produtividade (baixos H/M²).

## !!!!!Pontos importantes
- O gráfico de densidade mostra a concentração dos dados, destacando onde há mais valores de alta produtividade.

- O boxplot foca na mediana e na dispersão dos dados, que pode ser afetada por valores extremos (outliers) ou uma distribuição assimétrica.

```python
# Criando imports para a criação de gráficos de densidade
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")

# Calculando o índice de produtividade
df['Produtividade'] = df['COEF_SIURB'] / df['IP_D']

plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Produtividade', hue='OBRA', fill=False, common_norm=False, palette='viridis')
plt.title('Densidade de Produtividade por Obra')
plt.xlabel('Índice de Produtividade (H/M2)')
plt.ylabel('Densidade')
plt.show()
```

```python
produtividade_media = df.groupby('OBRA')['Produtividade'].mean().sort_values(ascending=False)

print("Índice de Produtividade por Obra:")
display(produtividade_media)

obra_menos_produtiva = produtividade_media.idxmax()
obra_mais_produtiva = produtividade_media.idxmin()

print(f"\nA obra mais produtiva é: {obra_mais_produtiva} com uma produtividade média de {produtividade_media.min():.4f} h/m²")
print(f"A obra menos produtiva é: {obra_menos_produtiva} com uma produtividade média de {produtividade_media.max():.4f} h/m²")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='OBRA', y='IP_D', data=df, hue='OBRA', palette='viridis', legend=False)
plt.title('Boxplot de Índice de Produtividade por Obra')
plt.xlabel('Obra')
plt.ylabel('Índice de Produtividade (H/M²)')
plt.show()
```



É importante deixar claro que o índice de produtividade medido, para os parametros solicitados, é o da fórmula: download.pngConforme a fórmula apresentada, é justificado o gráfico da seguinte forma:

- Na Obra A, podemos ver que sua área é muito grande, o que gera uma variaçao muito grande, e uma falta de noçao da estimativa de produtividade;

-  Na Obra B, analisamos que, a sua área é pequena, o que gera ótimas metas e expectativas, mas, por seus outliars extremamente inconstantes e diversos, dificilmente  se terá previsibilidade de resultados constantes, nao possuindo confiança nos dados.

- Na Obra C, temos um ótimo indice de confiabilidade, pelo pouquissimo numero de outliars, além de uma ótima area com a media de resultados.
